## Исследовательский анализ данных

Напишем функции для построения типовых графиков, в дальнейшем будем применять их.

```python
#напишем функцию для построения круговых диаграмм
def pie (data, column):
    values = data[column].value_counts()
    labels = values.index
    plt.figure(figsize=(5, 7))
    plt.pie(values, labels=labels, autopct='%.2f%%')
    plt.title(f'Распределение клиентов по признаку {column}', fontsize = 16, color = 'black')
    plt.show()
    
#напишем функцию для построения гистограмм
def hist (data, column):
    column_label = data.columns[data.columns.get_loc(column)]
    data[column].hist(figsize = (10,5))
    plt.xlabel(column_label, fontsize=12)
    plt.ylabel('Количество', fontsize=12)
    plt.title(f'Распределение клентов по {column}', fontsize = 16, color = 'black')
    plt.show()
    
#напишем функцию для построения ящика с усами
def box (data, column):
    plt.figure(figsize=(10, 2))
    sns.boxplot(data = data, x=column)
    plt.title(f'Распределение {column}')
    plt.ylabel(None)
    plt.xlabel(f'{column}', fontsize=12)
    plt.show()
    
#напишем функцию для построения столбчатой диаграммы
def bar (data, column, hue = None):
    column_label = data.columns[data.columns.get_loc(column)]
    plt.figure(figsize=(10, 5))
    sns.countplot(data=data, x=column, hue=hue)
    plt.xlabel(column_label, fontsize=12)
    plt.ylabel('Количество', fontsize=12)
    plt.title(f'Распределение клентов по {column}', fontsize = 16, color = 'black')
    plt.show()
 
```

### Анализ корреляции

Посмотрим, какие признаки клиентов имеют взаимосвязь. Для этого раздлим столбец с городом на отдельные столбцы и заменить значения в столбце gender на числовые.

```python
#заменим данные в столбце gender на числовые
bank = bank.replace({'gender': {'Ж':0, 'М': 1}})
#создадим числовой столбец
bank['city_int'] = bank.city.map({'Ярославль': 0, 'Рыбинск': 1, 'Ростов Великий': 2})
#разделим столбец с городами на отдельные колонки с числовыми значениями
bank = pd.get_dummies(bank, columns=['city_int'], drop_first= False)
#переименуем новые колонки с городами
bank = bank.rename(columns={"city_int_0": "city_yaroslavl", "city_int_1": "city_rybinsk", "city_int_2": "city_rostov"})
bank
```

```python
#построим матрицу корреляций
corr_matrix = bank.corr() 
#построим тепловую карту  для матрицы корреляций
plt.figure(figsize=(12,12))
sns.heatmap(corr_matrix, annot=True, square=True, cmap='GnBu', fmt='.0%')
plt.title('Тепловая карта по признакам')
plt.show()
```

<div class='alert alert-success'> 👍	
Круто что владеешь такой мощной техникой визуализации как тепловая карта. На ней прекрасно видно что особых корреляций между признаками нет. В такой ситуации стоит подумать о том, что возможно признаки имеют нелинейные связи. Для выявления таких нелинейных взаимосвязей я рекомендую дополнительно посмотреть на корреляцию Спирмена, а в идеале познакомиться с очень продвинутым коэффициентом корреляции Phik. С его помощью можно получать адекватные значения корреляции для данных любой природы. https://phik.readthedocs.io/en/latest/introduction.html


Небольшая прямая корреляция может быть отмечена между возрастом и уходом клиента (31%), а также между балансом и городом Ростовом Великим (41%). Существует также небольшая обратная корреляция между количеством продуктов и балансом(-33), а также между балансом и городом Ярославлем (-23%).  
В целом, можно сказать что взаимосвязи между показателями отсутствуют.

Возможно признаки имеют нелинейные связи. Для выявления таких нелинейных взаимосвязей воспользуемся коэффициентом корреляции Phik.

```python
#построим матрицу корреляций phik
bank_phik = bank.phik_matrix()
plt.figure(figsize=(12, 12))
sns.heatmap(bank_phik, annot = True, square=True, cmap= 'GnBu', fmt='.0%')
plt.title('Тепловая карта по признакам')

plt.show()
```

На тепловой карте phik мы видим связь оттока с возрастом (49%), с городом Ростовом (28%), с активностью(24%) и количеством продуктов (22%). Прослеживается взаимосвязь количества продуктов и баланса (54%), баланса и Ростова (45%), в общем-то и средний баланс в Ростове в два раза выше других городов.

## Статистическая проверка гипотез

<a id=5.1></a>
### Проверка гипотезы различия дохода между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним

Сформулируем нулевую и альтернативную гипотезу:  

Н0: различия в средних доходах между клиентами, которые пользуются 1 продуктом, и клиентами, которые пользуются 2 продуктами, нет  
Н1: различие в средних доходах между клиентами, которые пользуются 1 продуктом, и клиентами, которые пользуются 2 продуктами, имеется

```python
#создаем переменные для теста
product_1_est_sal = bank.query('products == 1')['estimated_salary']
product_2_est_sal = bank.query('products == 2')['estimated_salary']
```

```python
#проверяем данные
display(product_1_est_sal.describe())
display(product_2_est_sal.describe())
```

```python
print (stat.variance(product_1_est_sal))
print (stat.variance(product_2_est_sal))
```

Выборки достаточно большие и сопоставимые по показателям, в частности по среднему. Для анализа будем использовать 

```python
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(
    product_1_est_sal,
   product_2_est_sal, equal_var = False)

#задаем уровень статистической значимости
alpha = 0.05


print(f'P-value равен {round(results.pvalue, 3)}')

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Поскольку p-value нашего теста выше, чем установленный уровень статистической значимости, мы не можем отвергнуть нулевую гипотезу. У нас нет оснований, чтобы считать, что доходы клиентов, которые пользуются 1 продуктом, и клиентов, которые пользуются 2 продуктами, различаются.

### Проверка гипотезы различия в доходах между теми клиентами, которые уходят, и теми клиентами, которые остаются

Сформулируем нулевую и альтернативную гипотезу:  

Н0: различия в средних доходах между клиентами, которые уходят, и клиентами, остаются, нет  
Н1: различия в средних доходах между клиентами, которые уходят, и клиентами, остаются, имеются

```python
#создаем переменные для теста
churn_cliets_est_sal = bank.query('churn == 1')['estimated_salary']
non_churn_cliets_est_sal = bank.query('churn == 0')['estimated_salary']
```

```python
#проверяем данные
display(churn_cliets_est_sal.describe())
display(non_churn_cliets_est_sal.describe())
```

```python
print (stat.variance(churn_cliets_est_sal))
print (stat.variance(non_churn_cliets_est_sal))
```

```python
# results = вызов метода для проверки гипотезы
results = st.ttest_ind(
    churn_cliets_est_sal,
    non_churn_cliets_est_sal, equal_var = False)

#задаем уровень статистической значимости
alpha = 0.05

# вывод значения p-value на экран 
print(f'P-value равен {round(results.pvalue, 3)}')
# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Поскольку p-value нашего теста выше, чем установленный уровень статистической значимости, мы не можем отвергнуть нулевую гипотезу. У нас нет оснований, чтобы считать, что доходы клиентов, которые уходят, и клиентов, которые остаются, различаются.

### Проверка гипотезы различия оттока клиентов из головного офиса банка в Ярославле и из филиала в Ростове 

Сформулируем нулевую и альтернативную гипотезу:  

Н0: отток клиентов в Ростове Великом не отличается от оттока клиентов в Ярославле  
Н1: отток клиентов в Ростове Великом отличается от оттока клиентов в Ярославле

Поскольку в проверке будут участвовать бинарные значения будем примениять ztest.

```python
sample_size_a = bank.query('city == "Ярославль"').city.count()
sample_success_a = bank.query('city == "Ярославль" and churn == 1').city.count()
print(sample_size_a, sample_success_a)
```

```python
sample_size_b = bank.query('city == "Ростов Великий"').city.count()
sample_success_b = bank.query('city == "Ростов Великий" and churn == 1').city.count()
print(sample_size_b, sample_success_b)
```

```python
#задаем уровень статистической значимости
alpha = 0.05

successes = np.array([sample_success_a, sample_success_b])
samples = np.array([sample_size_a, sample_size_b])

stat, p_value = proportions_ztest(count=successes, nobs=samples,  alternative='two-sided')

print(f'z_stat: {stat} , p_value: {p_value}')

if p_value < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Поскольку p-value нашего теста ниже, чем установленный уровень статистической значимости, мы отвергли нулевую гипотезу. Есть основания считать, отток клиентов в Ростове Великом отличается от оттока клиентов в Ярославле.

Так как большая часть клиентов этого сегмента из Ярославля - развитого областного центра, а уровень баланса самый низкий (траты вероятнее больше, чем в городах уровнем поменьше), то можно предложить клиенту условия, позволяющие получать наибольшие выгоды от трат в городе: кэшбэк по широкому списку партнеров, возможно не кредитные, а карты рассрочки, топливные программы. Также, поскольку баланс данного этого клиента наиболее высок, предложить ему продукт, позволяющий сохранять и приумножать доход, возможно какую-нибудь инвестиционную программу.

### Портрет и рекомендации по клиентам из сегмента с 2 банковским продуктом

Вторая по численности группа. Самая стабильная и преданная банку, доля уходящих 7,4 %.  
Типичный клиент - мужчина 37 лет из Рыбинска, активный клиент, имеет кредитную карту, самый низкий баланс на счету - 51 401 у.е. и средний оцененный доход 100 410 у.е. Кредитный балл - 652, позволяет стабильно получать одобрение по кредиту.
Вероятно клиенты данного сегмента активные пользователи кредитной карты, имеют ипотеку или большой кредит, которые крепко держат клиента.

Клиенты данного сегмента скорее всего семейные люди с детьми из небольшого города рядом с административным центром. Для поддержания преданности банку таких клиентов можно предложить им выгодные вклады, льготные условия ипотеки и автокредитования.

### Портрет и рекомендации по клиентам из сегмента с 3 банковским продуктом

Самая малочисленная группа и самая уходящая группа клиентов (82,6 %).  
Типичный клиент - женщина среднего возраста (44 года) из Ростова Великого, неактивный пользователь, имеет кредитную карту, чаще всего перестающая быть клиентом банка. Оцененный доход по сравнению с другими сегментами самый высокий - 104 149 у.е., баланс на счете - 75 008 у.е. Кредитный балл  - 651, позволяет стабильно получать одобрение по кредиту.

В первую очередь, необходимо разобраться, какой продукт так сильно отталкивает клиента.  
А также необходимо разобраться, что происходит с клиентами в Ростове.  
Для поддрежания интереса к банку и повышения активности данного сегмента, необходимо сделать так, чтобы использование трех продуктов было инесло дополнительные выгоды. Вероятно можно делать предложения как премиум-клиенту, так как средний доход клиента данного сегмента выше всех, сформировать условия программы-премиум.
Так как большая часть клиентов - женщины, можно сделать карты с повышенным кэшбэком, например, в салонах красоты, магазинах одежды, косметики, украшений, фитнесе. Или может быть с возможностью самостоятельно выбрать категории.  Мужчинам, соответственно, можно предложить другие категории, например, автомобили, топливо, спорт.  
Так как клиенты с высоким доходом, для них можно предусмотреть различные инвестиционные продукты.

### Общие рекомендации

1. Провести исследования по уровню удовлетворенности обслуживанием по сегментам и по городам отельно, особо внимание обратить на Ростов Великий.
2. Проверить неточности в разрезе клиентов с 1 продуктом по одновременному наличию счета и кредитной карты.
3. Уточнить данные по пропускам в балансах, возможно нужно скорректировать ведение базы данных или исправить идущую ошибку. Пропусков в балансах нет в филиале в Ростове Великом.
4. Уточнить ведение столбца с объектами собственности, настораживают слишком большое количество объектов. Или тогда уточнить, что подпадает по объекты собственности, что из них значимо для банка. Либо это специфика клиентов, тогда предлагать продукты, связанные с собственность, например страхование.
Возможно стоит сравнить тарифы, количество банкоматов, удобство расположения офисов по городам.
